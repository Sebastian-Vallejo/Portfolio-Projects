```python
df2 = df.loc[
    
    (df['Event distance/length'].isin(['50km', '50mi'])) 
    &
    (df['Year of event'] == 2020) 
    & 
    (df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0)  == 'USA')
]
```

```python
#Clean up athlete age
```

```python
df2["athlete_age"] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
df2["Athlete performance"].str.split('h').str.get(0)
```

```python
df2["Athlete performance"] = df2["Athlete performance"].str.split('h').str.get(0)
```

```python
df2.head()
```

```python
#drop columns: Athlete club, Athlete Country, Athlete year of birth, Athlete age category

df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1, inplace =True)
```

```python
df2.head()
```

```python
#Clean up null values
```

```python
df2.isna().sum()
```

```python
df2.loc[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for dupes
df2.loc[df2.duplicated() == True]
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
#fix types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.head()
```

```python
#rename columns
```

```python
df2 = df2.rename(columns = {
    
'Year of event'              : 'year',                 
'Event dates'                : 'race_day',          
'Event name'                 : 'race_name',
'Event distance/length'      : 'race_length', 
'Event number of finishers'  : 'race_number_of_finishers',    
'Athlete performance'        : 'athlete_performance',
'Athlete gender'             : 'athlete_gender',    
'Athlete average speed'      : 'athlete_average_speed', 
'Athlete ID'                 : 'athlete_id'      
})
```

```python
#questions we may want to find out from the data
```

```python
#difference in speed for the 50k and 50 mi from both male and female
```

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are best in the 50M race (20 + Race minimum)(show 15)
```

```python
df3.loc[df3['race_length'] == "50mi"].groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count >19')
```

```python
#what age groups are worst in the 50M race (10 + Race minimum)(show 20)
```

```python
df3.loc[df3['race_length'] == "50mi"].groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count >10')
```

```python
#are runners slower in winter?
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.sample(15)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.sample(15)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
